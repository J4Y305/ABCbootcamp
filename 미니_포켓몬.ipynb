```python
labels = []
for i in df:
    labels.append(i)

colors = sns.color_palette('hls',len(labels))
fig = plt.figure(figsize=(8,8)) ## ìº”ë²„ìŠ¤ ìƒì„±
fig.set_facecolor('white')
ax = fig.add_subplot()


plt.title("Pokemon Type")
# startangle:ì‹œìž‘ê°ë„,  autopct:ë¶€ì±„ê¼´ì•ˆì— í‘œì‹œë˜ëŠ” ìˆ«ìž í˜•ì‹

pie = ax.pie(df["type"].value_counts(), ## íŒŒì´ì°¨íŠ¸ ì¶œë ¥
       startangle=90, ## ì‹œìž‘ì ì„ 90ë„(degree)ë¡œ ì§€ì •
       labels=df["type"].unique(),
       counterclock=False, ## ì‹œê³„ ë°©í–¥ìœ¼ë¡œ ê·¸ë¦°ë‹¤.
       autopct=lambda p : '{:.2f}%'.format(p), ## í¼ì„¼í‹°ì§€ ì¶œë ¥
       colors = colors ## ìƒ‰ìƒ ì§€ì •
       )
plt.axis('equal') # ì›ì˜ í˜•íƒœ ìœ ì§€
plt.show()
```
