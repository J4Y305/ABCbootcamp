```python
# 본체
from konlpy.tag import Komoran
from jamotools import split_syllables
from gensim.models import Word2Vec
import csv
import numpy as np
import pandas as pd

# Define the function to convert a sentence to a vector
def sentence_vector(sentence, model):
    words = komoran.morphs(sentence)
    jamo_words = [split_syllables(word) for word in words]
    word_vectors = [model.wv[jamo_word] for jamo_word in jamo_words if jamo_word in model.wv]
    if len(word_vectors) > 0:
        sentence_vector = sum(word_vectors)
        return sentence_vector / len(word_vectors)
    else:
        return None

# Load the DataFrame
df = pd.read_csv('500.csv', encoding='utf-8-sig')

# Initialize the morphological analyzer
komoran = Komoran()

# Preprocess the sentences for Word2Vec
corpus = []
sentences = df['review']
for sentence in sentences:
    # Morphological analysis
    words = komoran.morphs(sentence)
    
    # Split into jamo characters
    jamo_words = [split_syllables(word) for word in words]
    
    # Convert to the Word2Vec input format
    corpus.append(jamo_words)
    
# Train the Word2Vec model
model = Word2Vec(corpus, vector_size=1000, window=5, min_count=5, workers=4)

# Convert the sentences to vectors
vectors = np.zeros((len(df), model.vector_size))
for i, sentence in enumerate(sentences):
    vector = sentence_vector(sentence, model)
    if vector is not None:
        vectors[i, :] = vector

# Save the vectors to a CSV file
df['vec'] = vectors.tolist()
df.to_csv('vec_final_please.csv', index=False)
```
